## 📖 개요

> **MAF(Microsoft Agent Framework)** 는 .NET 및 Python 환경에서 AI 에이전트와 다중 에이전트 워크플로우를 구축하기 위한 오픈소스 개발 키트입니다.

### 🎯 핵심 특징

MAF는 **Semantic Kernel**과 **AutoGen** 프로젝트의 아이디어를 통합하고 확장하여, 두 프로젝트의 강점을 결합하고 새로운 기능을 추가했습니다. 동일한 개발팀에서 개발하고 있어 향후 AI 에이전트 구축을 위한 **통합 기반**을 제공할 것입니다.

## 🧩 핵심 구성 요소
MAF 프레임워크는 모델 클라이언트(chat completion 및 response), 상태 관리를 위한 에이전트 스레드, 에이전트 메모리를 위한 컨텍스트 제공자, 에이전트 동작을 가로채는 미들웨어, 그리고 도구 통합을 위한 MCP 클라이언트 등 기본적인 빌딩 블록 요소들을 제공합니다. **💡 이러한 구성요소들을 통해 상호작용적이고, 견고하며, 안전한 AI 애플리케이션을 구축할 수 있습니다.**

## 🤖 AI Agents란?

> **AI 에이전트는 LLM을 활용하여 자율적으로 작업을 수행하는 지능형 컴포넌트입니다.**

### 🔍 핵심 역할

AI 에이전트는 다음과 같은 작업을 수행합니다:

1. **입력 처리** - 사용자 입력을 이해하고 분석
2. **의사 결정** - LLM을 통한 지능적 판단
3. **도구 호출** - 도구 및 MCP 서버를 활용한 작업 수행
4. **응답 생성** - 결과를 바탕으로 적절한 응답 제공

### 🔧 확장 구성 요소

AI 에이전트는 다음 구성요소를 통해 기능을 향상시킬 수 있습니다:

- **스레드** (Threads) - 대화 상태 관리
- **컨텍스트 제공자** (Context Providers) - 메모리 및 컨텍스트 관리
- **미들웨어** (Middleware) - 실행 흐름 제어 및 확장

## 🔄 Workflow란?

> **워크플로우는 AI 에이전트를 포함한 미리 정의된 작업들의 시퀀스로, 일관성과 신뢰성을 보장합니다.**

### 🎯 설계 목적

워크플로우는 다음과 같은 복잡한 시나리오를 처리하도록 설계되었습니다:

- 🔀 **멀티 에이전트** - 여러 에이전트 간 협업
- 👥 **Human-in-the-loop** - 사람과의 상호작용
- 🔌 **외부 통합** - 외부 시스템과의 연동
- ⏱️ **장기 실행** - 복잡하고 오래 걸리는 프로세스

### 📊 워크플로우 실행 제어

워크플로우의 실행 시퀀스는 명시적으로 정의되어 실행 경로에 대한 세밀한 제어를 제공합니다.

### ⚡ 동적 시퀀스 지원

워크플로우는 다음 기능을 통해 동적 시퀀스를 표현할 수 있습니다:

| 기능 | 설명 |
|-----|------|
| **조건부 라우팅** | 조건에 따른 실행 경로 선택 |
| **모델 기반 의사 결정** | LLM을 활용한 지능적 판단 |
| **동시 실행** | 여러 작업의 병렬 처리 |

### 🎭 멀티-에이전트 오케스트레이션

이러한 동적 기능을 통해 **멀티-에이전트 오케스트레이션 패턴**이 구현됩니다. 오케스트레이션 패턴은:

- 여러 에이전트를 조정하여 복잡한 작업 수행
- 단일 에이전트의 한계 극복
- 협업을 통한 시너지 효과 창출
